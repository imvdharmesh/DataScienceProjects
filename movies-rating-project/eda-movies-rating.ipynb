# Movies Ratings - Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Movie-Ratings.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
#Q1)  list of top 10 movie(descending order) in terms of budget
df.sort_values(by='Budget (million $)',ascending=False).head(10)
```

```python
#Q2) list of top 10 movie (ascending order) in terms of budget
df.sort_values(by='Budget (million $)',ascending=True).tail(10)
```

```python
#Q3) which genre is the best in terms of price budget
df[['Budget (million $)','Genre']].max()
```

```python
sns.set_style('whitegrid')
```

```python
#Q4) distribution(graph) of audience rating
sns.distplot(df['Audience Ratings %'],color='red')
```

```python
#Q5) distribution(graph) of critics rating (rotten tomato)
sns.distplot(df['Rotten Tomatoes Ratings %'])
```

```python
#Q6) distribution of budget
sns.distplot(df['Budget (million $)'])
```

```python
#Q7) distribution of critics rating
sns.distplot(df['Rotten Tomatoes Ratings %'],color='green')
```

```python
#Q8) distribution of audience rating
sns.distplot(df['Audience Ratings %'])
```

```python
#Q9) histogram of critics rating 
plt.hist('Rotten Tomatoes Ratings %',data=df,bins=50,color='blue')
plt.xlabel('Rotten Tomatoes Ratings %')
plt.show()
```

```python
#Q10) histogram of audience rating
plt.hist('Audience Ratings %',bins=20,data=df)
plt.xlabel('Audience Ratings %')
plt.show()
```

```python
#Q11) show the total count of all the different genre present in the dataset
df['Genre'].value_counts()
```

```python
#Q12) show the graph of rating based on genre
sns.barplot('Genre','Audience Ratings %',data=df,palette='rainbow')
```

```python
#Q13) scatter plot of critics rating vs audience rating
sns.scatterplot('Rotten Tomatoes Ratings %','Audience Ratings %',data=df)
```

```python
#Q14) box plot based on genre and critics rating
sns.boxplot('Genre','Rotten Tomatoes Ratings %',data=df,palette='rainbow')
```

```python
#Q15) box plot based on genre and audience rating
sns.boxplot('Genre','Audience Ratings %',data=df,palette='coolwarm')
```

```python
#Q16) pairplot of the whole data set
sns.pairplot(df)
```

```python
#Q17)  histogram of  top 10 movie(descending order) in terms of budget
data1=df.sort_values(by='Budget (million $)',ascending=False).head(10)
data1[['Budget (million $)','Film']].hist()
plt.show()
```

```python
#Q18)  histogram of  top 10 movie(ascending order) in terms of budget
data2=df.sort_values(by='Budget (million $)',ascending=True).tail(10)
data2[['Budget (million $)','Film']].hist()
plt.show()
```

```python
#Q19) Joint Plot of critics rating and audience rating (with type as hexagonal)
sns.jointplot(x='Rotten Tomatoes Ratings %',y='Audience Ratings %',data=df,kind='hex')
```

```python
#Q20) Joint plot of budget vs genre (type as scatter)
sns.jointplot('Genre','Budget (million $)',data=df,kind='scatter')
```
